***Calculos matriciales***

```python
import numpy as np

A = np.array([[1,2], [3,4]])
B = np.array([[5,6], [7,8]])

print(A)
print(B)
print()

# SUMA
sum = A + B
print(sum)

# RESTA
resta = A - B
print(resta)
```

```python
# MULTIPLICACION
product = np.dot(A, B)
print(product)
```

```python
# DETERMINANTE
A = np.array([[1,2], [3,4]])
det_A = np.linalg.det(A).astype(np.float32)
print(det_A)
print()
inv_A = np.linalg.inv(A).astype(np.float32)
print(inv_A)
```

*Resolver Ax = b*

```python
b = np.array([9,11])
x = np.linalg.solve(A,b)
print(x)
```
