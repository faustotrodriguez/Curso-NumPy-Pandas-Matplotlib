```python
import numpy as np
array = np.array([[1,2,3],[4,5,6]])
print(array.ndim)
print(array.shape)
print(array.dtype)
```

```python
z = np.array(3, dtype=np.uint8)
print(z)
```

```python
double_array = np.array([1,2,3], dtype='d')
# dtype='d' equivale a dtype=np.float64
print(double_array)
```

```python
z = z.astype(np.float64)
print(z)
```

```python
array = np.array([[1,2,3],[4,5,6]])
```

```python
sum = np.sum(array)
mean = np.mean(array)
std = np.std(array)
print(sum, mean, std)
```
