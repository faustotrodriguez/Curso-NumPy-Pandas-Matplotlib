```python
import numpy as np

survey_responses = np.array(["bueno", "excelente", "malo", "bueno",
                              "bueno", "malo", "malo", "bueno",
                              "excelente", "malo"])

print(np.unique(survey_responses))
```

```python
unique_elements, counts = np.unique(survey_responses, return_counts=True)
print(unique_elements, counts)
```

```python
# Slicing básico (View)

array_1 = np.arange(10)
view_1 = array_1[1:3]

print(array_1)
print(view_1)

array_1[1:3] = [10,11]
print(array_1)
print(view_1)
```

```python
# Fancy Indexing (Copy)

array_x = np.arange(5, 10)
copy_x = array_x[[1,2]]

print(array_x)
print(copy_x)

array_x[1:3] = [10, 11]

print(array_x)
print(copy_x)
```

```python
# Cómo saber si es un Copy o un View

print(view_1.base is array_1) # Devuelve True si es una vista del original
print(copy_x.base is array_x)

print(view_1.base is None) 
print(copy_x.base is None) # Devuelve True si es un objeto independiente
```
