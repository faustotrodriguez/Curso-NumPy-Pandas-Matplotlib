```python
import pandas as pd

path = '/home/eltito/repositorios/Curso-NumPy-Pandas-Matplotlib/Online_Retail.csv'
df = pd.read_csv(path, encoding='latin1')
```

```python
country_count = df['Country'].value_counts()

# groupby = agrupame por 'Country', la sum() de 'Quantity'
country_group = df.groupby('Country')['Quantity'].sum()
country_group
```

```python
country_stats = df.groupby('Country')['UnitPrice'].agg(MeanPrice = 'mean', SumPrice = 'sum')
country_stats
```

```python
country_stock_group = df.groupby(['Country', 'StockCode'])['Quantity'].agg(QuantitySum = 'sum')
country_stock_group
```

```python
def total_revenue(group):
    return (group['Quantity'] * group['UnitPrice']).sum()

revenue_per_country = df.groupby('Country').apply(total_revenue)
revenue_per_country
```

```python
country_rank_bests = revenue_per_country.sort_values(ascending=False)
top3_best = country_rank_bests.iloc[:3]

country_rank_worsts = revenue_per_country.sort_values()
top3_worst = country_rank_worsts.iloc[:3]

print(top3_best)
print()
print(top3_worst)
```
