```python
import pandas as pd

path = '/home/eltito/repositorios/Curso-NumPy-Pandas-Matplotlib/Online_Retail.csv'
df = pd.read_csv(path, encoding='latin1')

# Esto es una manipulaciÃ³n, puedo crear una columna nueva a partir de otras
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df.head()
```

```python
df['HighValue'] = df['TotalPrice'] > 100

# La columna HighValue trae True si TotalPrice es > 100
df['HighValue'].value_counts()
```

```python
# Con esto traigo informaciÃ³n sobre el df
df.info()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.info()
```

```python
df['DiscountedPrice'] = df['UnitPrice'].apply(lambda x: x*0.9)
df.head(3)
```

```python
# Una funcion de categorizaciÃ³n que se aplica a una columna
def categorize_price(price):
    if price > 50:
        return 'High'
    elif price >= 20:
        return 'Medium'
    else:
        return 'Low'

df['PriceCategory'] = df['UnitPrice'].apply(categorize_price)

# Puedo usar esta columna creada y contar cuantos son por cada valor
df['PriceCategory'].value_counts()
```
