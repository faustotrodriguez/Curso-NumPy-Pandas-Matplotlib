```python
import pandas as pd
import numpy as np

path = '/home/eltito/repositorios/Curso-NumPy-Pandas-Matplotlib/Online_Retail.csv'
df = pd.read_csv(path, encoding='latin1')
```

```python
# Convertir la columna InvoiceDate a datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Eliminar filas con valores faltante en columnas críticas
df = df.dropna(subset= ['CustomerID', 'InvoiceDate'])

# Crear una nueva columna 'TotalPrice'
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

df.head(2)
```

```python
# Filtrar ventas en el Reino Unido - UK

uk_sales = (df['Country'] == 'United Kingdom').value_counts()

uk_sales2 = (df['Country'] == 'United Kingdom').sum()

uk_sales2
```

```python
# Filtrar ventas por cantidad mayor a 10

high_quality_sales = df[df['Quantity'] > 10]
high_quality_sales
```

```python
# Combinamos, ventas que sean mayor a 10 y que sean de United Kingdom

uk_high_quantity_sales = df[(df['Country'] == 'United Kingdom') & (df['Quantity'] > 10)]
uk_high_quantity_sales
```

```python
# Filtración por año

sales_2010 = df[df['InvoiceDate'].dt.year == 2010]
sales_2010
```

```python
# Filtración combinada por mes y año

sales_dec_2010 = df[(df['InvoiceDate'].dt.year == 2010) & (df['InvoiceDate'].dt.month == 12)]
sales_dec_2010
```
