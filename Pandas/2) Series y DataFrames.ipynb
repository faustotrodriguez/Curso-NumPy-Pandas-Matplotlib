```python
import pandas as pd, numpy as np

path = '/home/eltito/repositorios/Curso-NumPy-Pandas-Matplotlib/Online_Retail.csv'
retail_data = pd.read_csv(path, encoding='latin1')

print(retail_data)
```

```python
# Nombres de las columnas

columns_names = retail_data.columns
print(columns_names)
```

```python
# DimensiÃ³n

num_rows, num_columns = retail_data.shape
print(f"Filas: {num_rows}, columnas: {num_columns}")
```

```python
# Pongo en una variable los valores de una columna

daily_sales = retail_data['Quantity']
print(daily_sales)
print()

# Puedo ver el valor de una fila exacta

print(daily_sales[541904])
```

```python
# describe() me da un contexto sobre todas las columnas numÃ©ricas

summary = retail_data.describe().astype('Float32')
print(summary)
```

```python
daily_sales = retail_data['Quantity']

mean_value = daily_sales.mean()
print(f"La media es: {mean_value}")

median_value = daily_sales.median()
print(f"La mediana es : {median_value}")

sum_value = daily_sales.sum()
print(f"La suma total es: {sum_value}")

# count cuenta no nulos

count_values = daily_sales.count()
print(f"Conteo: {count_values}")
```
