```python
import pandas as pd, numpy as np

path = '/home/eltito/repositorios/Curso-NumPy-Pandas-Matplotlib/Online_Retail.csv'
retail_data = pd.read_csv(path, encoding='latin1')
```

```python
# Busqueda de datos faltantes

missing_data = retail_data.isna()
missing_data

missing_data_count = retail_data.isna().sum()
missing_data_count
```

```python
# Puedo eliminar las filas con datos faltantes

no_missing_rows = retail_data.dropna()
no_missing_rows
```

```python
# o elminar las columnas

no_missing_columns = retail_data.dropna(axis=1)
no_missing_columns
```

***Pero borrar filas o columnas no siempre es necesario***

```python
# Puedo llenar los datos faltantes con fillna

retail_data_filled_zeros = retail_data.fillna(0)
missing_retail_data_filled_zeros = retail_data_filled_zeros.isna().sum()
missing_retail_data_filled_zeros
```

```python
mean_unit_price = retail_data['UnitPrice'].mean()

retail_data_filled_mean = retail_data['UnitPrice'].fillna(mean_unit_price)
```
